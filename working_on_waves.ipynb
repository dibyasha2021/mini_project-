Working with multiple waves

```python
# used to create block diagrams

    
import numpy as np                   # for multi-dimensional containers
import pandas as pd                  # for DataFrames
import plotly.graph_objects as go    # for data visualisation

from plotly.subplots import make_subplots
```

```python
# creating Multiple Sine waves
sample_rate = 1000
start_time = 0
end_time = 10
theta = 0

time = np.arange(start_time, end_time, 1/sample_rate)
```

```python
# Creating list of frequencies and amplitude
frequency = [3, 5, 2, 1, 10]
amplitude = [1, 2, 7, 3, 0.1]
```

```python
# five frequency/amplitude values find sine waves
#Here frequency and Amplitude changes  each time

fig = make_subplots(rows=5, cols=1, shared_xaxes=True)

for i in range(5):
    sinewave = amplitude[i] * np.sin(2 * np.pi * frequency[i] * time + theta)  
    # theta remains 0 
    
    fig.add_scatter(x=time, y=sinewave, row=i+1, col=1, name=f"wave {i+1}")
    
fig.show()
```

```python
#Merging these sine waves
sinewave = np.empty(len(time)) #The empty() function is used to create a new array of given shape and type, without initializing entries.


for i in range(5):
    sinewave += amplitude[i] * np.sin(2 * np.pi * frequency[i] * time + theta)
```

```python
#figure
fig = go.Figure(layout=dict(xaxis=dict(title='Time (sec)'),yaxis=dict(title='Amplitude')))
fig.add_scatter(x=time, y=sinewave )
fig.show()
```
